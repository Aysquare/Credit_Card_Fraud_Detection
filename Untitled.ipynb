### Under-sampling

building a sample dataset containing similar distribution of normal transaction and fraudulent transaction

```python
legit_sample = legit.sample(n = 492)
```

```python
# Concateneting two dataframe

new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
#Investingating the first 5 rows of the balanced dataset

new_dataset.head()
```

```python
# Investigating the shape of the balanced dataset

new_dataset.shape
```

```python
# Confirming the distribution of the balanced dataset

new_dataset['Class'].value_counts()
```

```python
# Comparing the mean value of the class variable against other columns

new_dataset.groupby('Class').mean()
```

### Splitting the data into features and target



```python
X = new_dataset.drop(['Class'], axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

### Splitting the data into train and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### Model Training

LogisticRegression

```python
model = LogisticRegression()

# Training the logistic model with train data
model.fit(X_train,Y_train)
```

### Model Evaluation

Training data

```python
train_data_pred = model.predict(X_train)

accuracy = accuracy_score(Y_train, train_data_pred)
print('Train data accuracy score: ', accuracy)
```

Test data

```python
test_data_pred = model.predict(X_test)

accuracy = accuracy_score(Y_test, test_data_pred)
print('Test data accuracy score: ', accuracy)
```
